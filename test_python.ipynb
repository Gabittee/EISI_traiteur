## Exercice 3 : tableau de bord

Pour avoir une vision synthétique des informations, une nouvelle version du programme est mise en oeuvre. Le responsable d'exploitation souhaite disposer d'un tableau de bord affichant la température moyenne mesurée dans la journée, la température maxi et la température mini.

Vous devrez obtenir l'affichage ci-dessous:
```
Tableau de bord
---------------------------------------------------------
                moy min max
chambrefroide01 3.4 2.5 6.4
chambrefroide02 2.8 2.5 4.1
chambrefroide03 2.7 2.5 3.6
```

````python
# affichage des temperatures v1
import csv

max = [0,0,0]
min = [100,100,100]
moy = [0,0,0]
nb_chambre = 3

# Affichage de l'entete du programme
print("Tableau de bord")
print('-'*57)
print("                moy min max")
for num_chambrefroide in range(nb_chambre):

    num_chambrefroide_str = f'{num_chambrefroide + 1}'
    recherche = 'chambrefroide0' + num_chambrefroide_str
    print(recherche,end=' ')   
    # Ouverture du fichier de données
    with open('traiteur.csv') as csv_file:
        contenu_csv = csv.reader(csv_file, delimiter=';')
        
        # parcours des données du fichier
        for champ in contenu_csv:
            if champ[2] == recherche:
                temperature_corrigee = champ[3].replace(',', '.')
                temperature = float(temperature_corrigee)

                # prends le max 
                if temperature > max[num_chambrefroide]:
                    max[num_chambrefroide] = temperature
                
                # prends le min
                if temperature < min[num_chambrefroide]:
                    min[num_chambrefroide] = temperature
                    
                # Ajoute les 24 résultats à cette variable
                moy[num_chambrefroide] = moy[num_chambrefroide] + temperature

    # fait la moyenne
    moy[num_chambrefroide] = moy[num_chambrefroide] / 24
    
    # écris les résultats
    print(round(moy[num_chambrefroide], 1),end=' ')
    print(min[num_chambrefroide],end=' ')
    print(max[num_chambrefroide])
````
